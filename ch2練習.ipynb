```python
top=float(input("請輸入梯形上底長度："))
bottom=float(input("請輸入梯形下底長度："))
height=float(input("請輸入梯形高度："))
area=(top+bottom)*height/2
print("梯形的面積："+str(area))
#計算梯形面積
```

```python
long=float(input("請輸入長度："))
wide=float(input("請輸入寬度："))
area=long*wide
print("長方形的面積："+str(area))
#計算長方形面積
```

```python
deposit=int(input("請輸入本金存款金額"))
times=1.02**6
deposit*=times
print("6年後的存款為；"+str(deposit))
#計算複利本金
```

```python
deposit=int(input("請輸入手機金額："))
residual=50000-deposit
print("剩餘款為："+str(residual))
#延伸練習
```

```python
radius=float(input("請輸入圓半徑"))
area=3.14*radius**2
circum=radius*2*3.14
print("圓面積為："+str(area))
print("圓周長為："+str(circum))
#這個錯誤是因為您嘗試將一個字串與一個浮點數直接使用 + 運算子串聯。
#在 Python 中，如果您想將字串和數字一起列印，需要先使用 str() 函數將數字轉換為字串。
```

```python
distance=int(input("請輸入路程公里數(整數)"))

if distance <= 1:
    cost = 85 # 第一公里固定費用
else:
    cost = 85 + (distance - 1) * 20 # 第一公里後每公里加收20元

print("你的車程車資費="+str(cost))
```

```python
height=int(input("請輸入你的身高(cm)："))
weight=int(input("請輸入你的體重(kg)："))
bmi=weight/(height/100)**2
print(f"你的BMI為 {bmi:.2f}")
# f-string 來控制浮點數的顯示精度，讓 BMI 計算結果只顯示到小數點後兩位
```

```python
#1英吋=2.54公分
#1英呎=12英吋
height=int(input("請輸入你的身高(cm)"))
inch=height/2.54
feet=round(inch//12)
inch=round(inch%12) # 將英吋數四捨五入為整數
print("身高"+str(height)+"公分等於"+str(feet)+"呎"+str(inch)+"吋")
```
