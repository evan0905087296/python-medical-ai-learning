```python
import torch

from monai.networks.nets import UNet
```

```python
model = UNet(
    spatial_dims=3,
    in_channels=4,
    out_channels=3,
    channels=(16,32,64),
    strides=(2,2),
    num_res_units=2
)
```

```python
print(model)
```

```python
image = torch.randn(
    1,
    4,
    240,
    240,
    155
)

print(image.shape)
```

```python
with torch.no_grad():

    output = model(image)
    print(output.shape)
```
